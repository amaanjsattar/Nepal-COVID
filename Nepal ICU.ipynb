```python
Nepal_today = Nepal.where('Date', '4/29/2021')
```

```python
Nepal_today
```

```python
Nepal_ICU_Ventilators = Nepal_today.select("Province", "ICU_Occupancy", "Ventilators_Occupancy")
```

```python
Nepal_ICU_Ventilators = Nepal_ICU_Ventilators.relabeled(["ICU_Occupancy", "Ventilators_Occupancy"], ["ICU", "Ventilators"])
```

```python
fig = Nepal_ICU_Ventilators.barh('Province', xaxis_title = 'Occupancy (%)',\
        title = 'ICU and Ventilator Occupancies in the Provinces of Nepal on April 29, 2021', show = False)
```

The dataset found on [Kaggle](https://www.kaggle.com/akashsky13/nepal-icu-and-ventilators-occupancy-for-covid) contains information from Nepal's Ministry of Health and Population's COVID briefing. The dataset includes information regarding ICU occupancy, total space, ventilator quantity, and ventilator availability for Nepal's seven provinces from April 12-29. 

I utilized data collected on April 29, reflecting the most recent counts for ICU and ventilator availability in each province. It appears Lumbini, Karnali, and Sudur-Paschim are the provinces running lowest on ICU and ventilator availability, while Gandaki has the most ICU and second-most ventilator occupancy. All of Pradesh-2s ventilators remain available on April 29, though 20% of the ICU is occupied. The dataset suggests that relief efforts would benefit from prioritizing Sudur-Paschim, Karnali, and Lumbini, the provinces with the least remaining ICU and ventilator occupancy.
